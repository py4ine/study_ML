```python
# 데이터 불러오기
import seaborn as sns

data = sns.load_dataset('penguins')
```

```python
# 데이터 확인
data
```

```python
data.info()
```

```python
# 데이터 전처리 NaN값 제거
penguins = data.dropna()
penguins = data.drop('species', axis=1)
penguins
```

# K-means

```python
# 라이브러리 불러오기 
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
```

```python
pd.get_dummies(penguins)
```

```python
penguins_arr = np.array(penguins)
penguins_arr
```

```python
# 모델생성(클러스터수3)
km = KMeans(n_clusters=3, random_state=42)

# 모델학습
km.fit(penguins_arr)
```
